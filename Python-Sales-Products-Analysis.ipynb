```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 # Step 1: Data Consolidation and Merging
```

```python
sales_part1=pd.read_csv("sales_part1.xls")
sales_part1.head()
```

```python
sales_part2=pd.read_csv("sales_part2.xls")
sales_part2.head()
```

```python
products=pd.read_csv("products.xls")
products.head()
```

```python
suppliers=pd.read_csv("suppliers.xls")
suppliers.head()
```

```python
all_sales=pd.concat([sales_part1,sales_part2],ignore_index=True)
all_sales.head()
```

```python
all_saless=pd.merge(all_sales,products,on="product_id",how="left")
all_saless.head()
```

```python
master_df=pd.merge(all_saless,suppliers,on="supplier_id",how="left")
master_df.head()
```

```python
master_df.tail()
```

```python
master_df.sample(2)
```

```python
master_df["revenue"]=master_df["quantity"] * master_df["sale_price"]
```

```python
master_df["profit"]=( master_df["sale_price"]- master_df["cost_price"]) * master_df["quantity"]
```

```python
master_df.head()
```

```python
Q1=master_df["quantity"].quantile(0.25)
Q3=master_df["quantity"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1 - 1.5 * IQR
upper_bound=Q3 + 1.5 * IQR
```

```python
print(f"Original dataset shape: {master_df.shape}")
```

```python
master_no_outliers_df=master_df[(master_df["quantity"]>=lower_bound) & (master_df["quantity"]<=upper_bound)]
```

```python
print(f"New dataset shape without outliers: {master_no_outliers_df.shape}")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=master_df,y="quantity",  x="category",palette='coolwarm')
plt.title("Quantity Distribution with Outliers")
plt.xlabel("Quantity") 
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=master_no_outliers_df,y="quantity",  x="category",palette='coolwarm')
plt.title("Quantity Distribution without Outliers")
plt.xlabel("Quantity") 
plt.ylabel("Category")
plt.show()
```

```python
# Step 3: Grouping and Summarizing Data
```

```python
category_saless=master_no_outliers_df.groupby("category").agg({
    "revenue":"sum","profit":"sum"})
category_saless
```

```python
category_sales=master_no_outliers_df.groupby("category")["revenue"].sum().sort_values(ascending=False).reset_index()
category_sales
```

```python
supplier_sales=master_no_outliers_df.groupby("supplier_name")["revenue"].sum().sort_values(ascending=False).reset_index()
```

```python
master_no_outliers_df["order_date"] = pd.to_datetime(master_no_outliers_df["order_date"])
master_no_outliers_df["Month"] = master_no_outliers_df["order_date"].dt.month_name()
master_no_outliers_df.sample(2)
```

```python
monthly_sales=master_no_outliers_df.groupby("Month")["revenue"].sum().sort_index().reset_index()
monthly_sales
```

```python
# Step 4: Data Visualization
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=category_sales,x="category",y="revenue",palette='viridis')
plt.title("Total Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=supplier_sales,y="supplier_name",x="revenue",palette='magma')
plt.title("Total Revenue by Supplier")
plt.xlabel("Revenue")
plt.ylabel("Supplier")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales,x="Month",y="revenue",marker="o")
plt.title("Total Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
